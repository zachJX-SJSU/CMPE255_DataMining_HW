```python
import pandas as pd
import seaborn as sns

# simple dataset for anomaly detection
df = sns.load_dataset("iris").drop(columns=["species"])
print("✅ Dataset loaded:", df.shape)
df.head()
```

```python
from pycaret.anomaly import *

exp_ano = setup(
    data=df,
    session_id=42,
    normalize=True,
    use_gpu=True,
    html=False
)
```

```python
iforest = create_model('iforest')
iforest
```

```python
anomalies = assign_model(iforest)
anomalies.head()
```

```python
plot_model(iforest, plot='tsne')
```

```python
knn_model = create_model('knn')
anomalies = assign_model(knn_model)
anomalies.head()

```

```python
plot_model(knn_model, plot='tsne')
```

```python
get_config('X_train_transformed').head()
print("Current seed:", get_config('seed'))
set_config('seed', 99)
print("New seed:", get_config('seed'))
```
